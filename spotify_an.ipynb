## User Demographics

```python
df.isnull().sum() #tocheck missing values
df.dropna(inplace=True) #drop if needed
df.columns = df.columns.str.strip().str.lower() #standardize column names
```

```python
df['age'].value_counts()
df['gender'].value_counts()
```

```python
sns.countplot(x='age', data=df)
plt.title("Age Distribution")
plt.show()
```

## Favourite Music Genre

```python
genre_counts = df['fav_music_genre'].value_counts().head(10)
genre_counts
```

```python
genre_counts.plot(kind='bar')
plt.title("Top Music Genres")
plt.xticks(rotation=45) #lables of x-axis
plt.show()
```

## Subscription Analysis

```python
sns.countplot(x='spotify_subscription_plan', data=df)
plt.title('Subscription Plan Distribution')
plt.xticks()
plt.show()
```

## Listening Time Behaviour

```python
df['music_time_slot'].value_counts()
```

```python
sns.countplot(x='music_time_slot', data=df)
plt.title("Preferred Listening Time")
plt.xticks()
plt.show()
```

## Recommendation System Analysis

# Insight Section

## Key Insights:
- Most users belong to 20-35 age group, indicating target audience concentration.
- Certain genres like Melody and Classical dominate listening preferences, showing clear user taste patterns.
- A significant portion of users use free plans, which could be an opportunity for premium conversion.
- Recommendation ratings suggest average satisfaction levels.
- Listening behavior peaks during night time slots, useful for targeted recommendations.

## Business Implications:
- Improve recommendation algorithms for low-rated users.
- Target ads/promotions during peak listening hours.
- Focus on popular genres for better engagement.
